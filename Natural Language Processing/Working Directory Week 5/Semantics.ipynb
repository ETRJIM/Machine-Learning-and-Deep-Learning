```python
from scipy.spatial.distance import cosine
cosine_similarity = lambda vec1,vec2: 1-cosine(vec1,vec2)
```

```python
king = nlp.vocab['king'].vector
man = nlp.vocab['man'].vector
queen = nlp.vocab['queen'].vector
new_vector = man-king+queen
```

```python
count = 0
iterations = 0
similarities = []
for s in nlp.vocab.vectors:
    x = nlp.vocab[s].text 
    if (x.islower() and x.isalpha() and len(x)>=3 and len(x)<=11):
        simi = cosine_similarity(new_vector,nlp.vocab[s].vector)
        similarities.append((simi,x))
        count += 1
    elif count>300000:
        break
    else:
        pass
    iterations += 1
print(iterations,count)
similarities = sorted(similarities)[::-1]
similarities[:10]
```
